### Restructuring the confirmed cases DataFrame

```python
africa = ['Algeria','Angola','Benin','Botswana','Burkina Faso','Burundi', 'Cabo Verde', 'Cameroon','Chad', 'Mozamibue','Congo (Brazzaville)',
          'Central African Republic', "Cote d'Ivoire", 'Congo (Kinshasa)','Egypt','Equatorial Guinea', 'Eritrea','Eswatini','Ethiopia','Ghana','Guinea',
          'Guinea-Bissau','Kenya','Liberia','Rwanda','Libya','Madagascar','Mauritius','Morocco','Mozambique','Namibia','Niger','Nigeria','Seychelles',
          'Tanzania','Togo','Sierra Leone', 'Malawi', 'South Sudan','Tunisia','Uganda','Senegal','South Africa','Sudan','Western Sahara','Zambia', 'Zimbabwe']
```

```python
confirmed = confirmed.drop(['Lat', 'Long'], axis=1)
ncountries = confirmed['Country/Region'].unique().tolist()

confirmed_cases = pd.DataFrame()
for i, country in enumerate(africa):#enumerate(ncountries):
    positive_cases = confirmed[confirmed['Country/Region'] == country].copy()
    if len(positive_cases) > 1:
        positive_cases = positive_cases.drop(['Province/State'], axis=1).groupby('Country/Region').sum().reset_index()
    else:
        positive_cases = positive_cases.drop(['Province/State'], axis=1)
        
    positive = positive_cases.melt(id_vars=["Country/Region"], 
                    var_name="Date",
                    value_name="confirmed_cases")
    positive = positive.rename({'Country/Region': 'country', 'Date': 'date'}, axis=1)
    confirmed_cases = confirmed_cases.append(positive )
    
confirmed_cases['date'] = pd.to_datetime(confirmed_cases['date'])
confirmed_cases.head()
```

```python
#dropping the days with 0 confirmed cases

confirmed_cases = confirmed_cases[confirmed_cases.confirmed_cases != 0]
confirmed_cases.head()
```

```python
#resetting indexes
confirmed_cases = confirmed_cases.reset_index(drop=True)
confirmed_cases.head()
```

### Restructuring the number of recoveries DataFrame

```python
recoveries = recoveries.drop(['Lat', 'Long'], axis=1)
ncountries = recoveries['Country/Region'].unique().tolist()

recovered = pd.DataFrame()
for i, country in enumerate(africa):#enumerate(ncountries):
    recover_no = recoveries[recoveries['Country/Region'] == country].copy()
    if len(recover_no) > 1:
        recover_no = recover_no.drop(['Province/State'], axis=1).groupby('Country/Region').sum().reset_index()
    else:
        recover_no = recover_no.drop(['Province/State'], axis=1)
        
    negetive = recover_no.melt(id_vars=["Country/Region"], 
                    var_name="Date",
                    value_name="number_recoveries")
    negetive = negetive.rename({'Country/Region': 'country', 'Date': 'date'}, axis=1)
    recoveries = recoveries.append(negetive )
    
recovered['date'] = pd.to_datetime(recovered['date'])
#recovered.head()
```
